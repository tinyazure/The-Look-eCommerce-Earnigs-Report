#2025Q2 Revenue Calculation

> Explore Products and Orders Table

```python
%%bigquery --project {project_id}
select
*
from
`bigquery-public-data.thelook_ecommerce.products`
```

```python
%%bigquery --project {project_id}
select
id as Product_Id,
COUNT(id) as Count_Id
from
`bigquery-public-data.thelook_ecommerce.products`
group by
id
having COUNT(id) >1
```

```python
%%bigquery --project {project_id}
select
  *
from
  `bigquery-public-data.thelook_ecommerce.order_items`
where
product_id=1
```

```python
%%bigquery --project {project_id}
select
  product_id,
  count(distinct sale_price) as distinct_sale_price_count
from
  `bigquery-public-data.thelook_ecommerce.order_items`
group by
  product_id
having distinct_sale_price_count >1
```

```python
%%bigquery --project {project_id}
select
status as status
from
`bigquery-public-data.thelook_ecommerce.order_items`
group by
status

#this mean that revenue = Complete + Shipped
```

```python
%%bigquery --project {project_id}
select
  OI.product_id,
  OI.order_id,
  round(sum(OI.sale_price),2) as Revenue,
from
  `bigquery-public-data.thelook_ecommerce.order_items` OI
group by
  OI.product_id,
  OI.order_id
order by
  product_id desc
```

```python
%%bigquery --project {project_id}
select
status as Order_Status,
count(distinct order_id) as count_order_id,
round(sum(OI.sale_price),2) as Revenue,
from
  `bigquery-public-data.thelook_ecommerce.order_items` OI
where
OI.created_at between '2025-03-01' and '2025-06-30' and
OI.status in ('Complete','Shipped')
group by
status

UNION ALL

select
'Total' as status,
count(distinct order_id) as count_order_id,
round(sum(OI.sale_price),2) as Revenue,
from
  `bigquery-public-data.thelook_ecommerce.order_items` OI
where
OI.created_at between '2025-03-01' and '2025-06-30' and
OI.status in ('Complete','Shipped')
```
