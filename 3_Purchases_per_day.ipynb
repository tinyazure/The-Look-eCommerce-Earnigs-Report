# Purchases per Day Query

>For the following Query we will need to work on the "order_items" table. The table contains creation day, order id and sale price.

```python
%%bigquery --project {project_id}
select
  created_at,
  order_id,
  sale_price
from
  `bigquery-public-data.thelook_ecommerce.order_items`
limit 5
```

```python
%%bigquery --project {project_id}
select
  COUNT(DISTINCT(date(created_at))) as count_days,
  COUNT(DISTINCT(order_id)) as count_orders,
  ROUND(COUNT(DISTINCT(order_id)) / COUNT(DISTINCT(DATE(created_at)))) as purchases_per_day
from
  `bigquery-public-data.thelook_ecommerce.order_items`
where
  created_at between '2025-03-01' and '2025-06-30'
```
